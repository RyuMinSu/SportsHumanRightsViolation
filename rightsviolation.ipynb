```python
#####경험유무
# def exp_ox(df, column_name):
#   cc = df[column_name].value_counts()
#   cc_false = [cc["없음"] + cc["무응답"]]
#   cc_false_pct = [((cc["없음"] + cc["무응답"])/len(df)*100).round(1)]
#   cc_true = [len(df) -(cc["없음"] + cc["무응답"])]
#   cc_true_pct = [((len(df) -(cc["없음"] + cc["무응답"]))/len(df)*100).round(1)]


#   cc_df = pd.DataFrame({"있음": cc_true,
#                         "있음%": cc_true_pct,
#                         "없음": cc_false,
#                         "없음%": cc_false_pct})
#   return cc_df


def exp_ox(df, column_name):
  cc = df[column_name].value_counts() #값들
  cc_false = cc["없음"]
  cc_nan = cc["무응답"]
  cc_true = len(df) - (cc_false+cc_nan)
  cc_total = len(df) - cc_nan

  cc_false_pct = (cc_false / cc_total*100).round(1)
  cc_true_pct = (cc_true / cc_total*100).round(1)

  cc_df = pd.DataFrame({"있음": [cc_true],
                        "있음%": [cc_true_pct],
                        "없음": [cc_false],
                        "없음%": [cc_false_pct],
                        "무응답": [cc_nan],
                        "합계": [cc_total]})
  return cc_df
  

###경험횟수
def make_exp(df, column_name):
  bb = df[column_name].value_counts()
  index = bb.index
  values = bb.values
  pct = (bb.values / (len(df) - bb["무응답"])*100).round(1)

  exp_df = pd.DataFrame({"구분": index,
                         "N": values,
                         "%": pct})
  return exp_df


#####횟수 있음만, 5개 컬럼
def make_true_df(df, column_name, column_num):
  cond1 = df[column_name] == "없음"
  true_df = df[~cond1].iloc[:,column_num:column_num+6]
  return true_df

#####응답자특성(성별-횟수)
def sex_times(df, column1, column2):
  cross_df = pd.crosstab(df[column1], df[column2], margins=True)
  if "매일" in cross_df.columns:
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]    
  else:
    cross_df["매일"] = 0 
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]
  
  if "한달에 1~2회" in cross_df.columns:
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]    
  else:
    cross_df["한달에 1~2회"] = 0 
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]

  return cross_df


#####응답자특성(학년-횟수)
def class_times(df, column1, column2):
  cross_df = pd.crosstab(data[column1], data[column2], margins=True)
  if "매일" in cross_df.columns:
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]
  else:
    cross_df["매일"] = 0 
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]

  if "한달에 1~2회" in cross_df.columns:
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]    
  else:
    cross_df["한달에 1~2회"] = 0 
    cross_df = cross_df[["All", "매일", "1주에 1~2회", "한달에 1~2회", "6개월에 1~2회", "1년에 1~2회", "없음"]]
  return cross_df


#####행위자 더미변환
def make_person_dummies(tru_df, column_name):
  columns = ["교수", "조교", "선배", "동기", "기타(아래 작성)", "무응답"]
  # columns = ["교수", "조교", "선배", "동기", "기타(아래 작성)"]
  zero_matrix = np.zeros((len(tru_df), len(columns)))
  dummies = pd.DataFrame(zero_matrix, columns=columns)
  for idx, value in enumerate(tru_df[column_name]):
    indecies = dummies.columns.get_indexer(value.split(", "))
    dummies.iloc[idx, indecies] = 1
  return dummies


#####실시자 빈도&비율
def person_num_pct(dummies_df):
  #실시자 빈도
  p1n = [dummies_df["교수"].sum()]
  p2n = [dummies_df["조교"].sum()]
  p3n = [dummies_df["선배"].sum()]
  p4n = [dummies_df["동기"].sum()]
  p5n = [dummies_df["기타(아래 작성)"].sum()]
  total_person = dummies_df["교수"].sum() + dummies_df["조교"].sum() +\
  dummies_df["선배"].sum() + dummies_df["동기"].sum() + dummies_df["기타(아래 작성)"].sum()
  p6n = [dummies_df["무응답"].sum()]

  #실시자 비율
  
  # p1p = [(dummies_df["교수"].sum()/(len(dummies_df) - dummies_df["무응답"].sum())*100).round(1)]
  # p2p = [(dummies_df["조교"].sum()/(len(dummies_df) - dummies_df["무응답"].sum())*100).round(1)]
  # p3p = [(dummies_df["선배"].sum()/(len(dummies_df) - dummies_df["무응답"].sum())*100).round(1)]
  # p4p = [(dummies_df["동기"].sum()/(len(dummies_df) - dummies_df["무응답"].sum())*100).round(1)]
  # p5p = [(dummies_df["기타(아래 작성)"].sum()/(len(dummies_df) - dummies_df["무응답"].sum())*100).round(1)]

  p1p = [(dummies_df["교수"].sum()/total_person*100).round(1)]
  p2p = [(dummies_df["조교"].sum()/total_person*100).round(1)]
  p3p = [(dummies_df["선배"].sum()/total_person*100).round(1)]
  p4p = [(dummies_df["동기"].sum()/total_person*100).round(1)]
  p5p = [(dummies_df["기타(아래 작성)"].sum()/total_person*100).round(1)]
  # p6p = [(dummies_df["무응답"].sum()/total_person*100).round(1)]

  #실시자 데이터프레임
  place_cal_df = pd.DataFrame({"교수N": p1n,
                              "교수%": p1p,
                              "조교N": p2n,
                              "조교%": p2p,
                              "선배N": p3n,
                              "선배%": p3p,
                              "동기N": p4n,
                              "동기%": p4p,
                              "기타(아래 작성)N": p5n,
                              "기타(아래 작성)%": p5p,
                              "무응답N": p6n})
                              # "무응답N": p6p })
  return place_cal_df


#####장소 더미변환
def make_place_dummies(tru_df, column_name):
  columns = ["생활관", "강의실", "경기장 또는 훈련장", "교수연구실", "무응답"]
  zero_matrix = np.zeros((len(tru_df), len(columns)))
  dummies = pd.DataFrame(zero_matrix, columns=columns)
  for idx, value in enumerate(tru_df[column_name]):
    indecies = dummies.columns.get_indexer(value.split(", "))
    dummies.iloc[idx, indecies] = 1
  return dummies


#####장소 빈도&비율
def place_num_pct(dummies_df):
  #장소 빈도
  p1n = [dummies_df["생활관"].sum()]
  p2n = [dummies_df["강의실"].sum()]
  p3n = [dummies_df["경기장 또는 훈련장"].sum()]
  p4n = [dummies_df["교수연구실"].sum()]
  p5n = [dummies_df["무응답"].sum()]
  total_place = dummies_df["생활관"].sum() + dummies_df["강의실"].sum() +\
  dummies_df["경기장 또는 훈련장"].sum() + dummies_df["교수연구실"].sum()

  #장소 비율
  p1p = [(dummies_df["생활관"].sum()/total_place*100).round(1)]
  p2p = [(dummies_df["강의실"].sum()/total_place*100).round(1)]
  p3p = [(dummies_df["경기장 또는 훈련장"].sum()/total_place*100).round(1)]
  p4p = [(dummies_df["교수연구실"].sum()/total_place*100).round(1)]
  # p5p = [(dummies_df["무응답"].sum()/total_place*100).round(1)]

  #장소 데이터프레임
  place_cal_df = pd.DataFrame({"생활관N": p1n,
                              "생활관%": p1p,
                              "강의실N": p2n,
                              "강의실%": p2p,
                              "경기장 또는 훈련장N": p3n,
                              "경기장 또는 훈련장%": p3p,
                              "교수연구실N": p4n,
                              "교수연구실%": p4p,
                              "무응답N": p5n})
                              # "무응답%": p5p})
  return place_cal_df


#####받은영향
def make_effect(tru_df, column_name):
  tru_columns = tru_df.columns
  tru_df2 = tru_df[tru_df[tru_columns[0]]!="무응답"]


  indexes = tru_df2[column_name].value_counts().index
  values = tru_df2[column_name].value_counts().values
  no_answer = tru_df2[column_name].value_counts()["무응답"]
  pct = (values / (len(tru_df2)-no_answer) * 100).round(1)

  reg_df = pd.DataFrame({"보기": indexes,
                       "N": values,
                       "%": pct})
  reg_df.set_index("보기")
  return reg_df
```
